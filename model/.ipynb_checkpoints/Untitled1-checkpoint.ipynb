```python
import numpy as np
import pandas as pd
import yfinance as yf
from math import floor,ceil,sqrt
import sys
import warnings
warnings.filterwarnings('ignore')
import datetime as dt
import tensorflow as tf
```

```python
seed = 42
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
stock_ticker = 'MSFT'
start_date = dt.datetime(2010, 1, 1)
end_date = dt.date.today()
```

```python
df = yf.download(stock_ticker, start = start_date, end = end_date, progress=False)
df
```

```python
df["Date"] = df.index
df = df[["Date", "Open", "High", "Low", "Close", "Volume"]]
df.reset_index(drop=True, inplace=True)
df.dropna(inplace=True)
```

```python
import matplotlib.pyplot as plt
```

```python
df['Date'] = pd.to_datetime(df.Date,format='%Y/%m/%d')
df.index = df['Date']
plt.figure(figsize=(16,8))
plt.plot(df['Close'], label='Close Price history',color='g')
plt.xlabel('Date',size=20)
plt.ylabel('Stock Price',size=20)
plt.title('Stock Price over the Years',size=25)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
```

```python
df_scaled = scaler.fit_transform(df[['Close']])
```

```python
x = []
y = []
window_size=50
train_split=0.75
data = df_scaled
```

```python
for i in range(window_size, len(data)):
    x.append(data[i-window_size:i, 0])
    y.append(data[i, 0])

x, y = np.array(x), np.array(y)
```

```python
len(x)
```

```python
train_size = int(len(x) * train_split)
xtrain, xtest = x[:train_size], x[train_size:]
ytrain, ytest = y[:train_size], y[train_size:]
```

```python
xtrain = np.reshape(xtrain, (xtrain.shape[0], xtrain.shape[1], 1))
xtest = np.reshape(xtest, (xtest.shape[0], xtest.shape[1], 1))
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout, Bidirectional
```

```python
model = Sequential()
model.add(Bidirectional(LSTM(units=100, return_sequences=True, input_shape=(50, 1))))
model.add(Dropout(0.2))
model.add(Bidirectional(LSTM(units=100, return_sequences=False)))
model.add(Dropout(0.2))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
from keras.callbacks import EarlyStopping
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
history = model.fit(xtrain, ytrain, epochs=20, batch_size=32, validation_split=0.2, callbacks=[early_stopping], verbose=1)
```

```python
predictions = model.predict(xtest)
```

```python
predictions = scaler.inverse_transform(predictions)
ytest = scaler.inverse_transform(ytest.reshape(-1, 1))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
```

```python
# Calculate RMSE
rmse = sqrt(mean_squared_error(ytest, predictions))
print(f"RMSE: {rmse}")
```

```python
# Calculate MAPE
mape = mean_absolute_percentage_error(ytest, predictions) * 100
print(f"MAPE: {mape}")
```

```python
plt.figure(figsize=(14, 7))
plt.plot(ytest, color='blue', label='Actual Stock Price')
plt.plot(predictions, color='red', label='Predicted Stock Price')
plt.title('Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```
